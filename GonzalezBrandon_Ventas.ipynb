# *Selección Inicial*

***Acceder y aislar columnas específicas por su nombre.***

```python
df['Customer_Name'] # Listar información de una columna en especifíco
```

```python
df[['Customer_Name','Revenue']] # Mostrar registros de dos columnas
```

***Seleccionar datos por sus etiquetas de fila y columna (nombres)***

```python
df.loc[0] # Información del registro 0
```

```python
df.loc[156] # Información del registro 156
```

```python
df.loc[4500] # Información del registro 4500
```

```python
revenue_year = df.loc[
    (df['Year'] == 2023) & (df['Product'] == 'Product A'), # Condición para las filas
    ['Month', 'Product', 'Units_Sold', 'Revenue']       # Lista de columnas a mostrar
]

print(revenue_year.head(10))
```

```python
product_month = df.loc[
    (df['Month'] == 'Feb'), # Condición para las filas
    ['Product', 'Price_per_Unit', 'Units_Sold', 'Revenue']       # Lista de columnas a mostrar
]

print(product_month.head(10))
```

```python
revenue_year_month = df.loc[
    (df['Year'] == 2021) & (df['Month'] == 'Jul') & (df['Product'] == 'Product C'), # Condición para las filas
    ['Year', 'Month', 'Product', 'Price_per_Unit', 'Units_Sold', 'Revenue']       # Lista de columnas a mostrar
]

print(revenue_year_month.head(10))
```
